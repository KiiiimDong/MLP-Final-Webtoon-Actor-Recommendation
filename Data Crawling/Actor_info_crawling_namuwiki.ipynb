# 나무위키 : 배우이름, 배우이름 링크, 배우나이

## 1. 패키지 로드 및 크롬 드라이버 설정

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm_notebook

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--lang=ko_KR')
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument('--window-size=1920x1080')

    # headless임을 숨기기 위해서
    # headless인 경우 Cloudflare 서비스가 동작한다.
    chrome_options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.53 Safari/537.36")

    driver = webdriver.Chrome(service=Service(
        ChromeDriverManager().install()), options=chrome_options)
    return driver
```

## 2. 배우이름, 배우이름링크

```python
driver = set_chrome_driver()
```

```python
url = 'https://namu.wiki/w/%EB%B0%B0%EC%9A%B0/%ED%95%9C%EA%B5%AD' # 나무위키 : 한국/배우
driver.get(url=url)
html = driver.page_source
soup = BeautifulSoup(html)
```

```python
actors_area = soup.select('a._0BBV4ix3')
actors_area
```

```python
actors_dict = [{'Name' : i.get('title'),
               'Name_url' : 'https://namu.wiki'+i.get('href')} for i in actors_area]
```

```python
del actors_dict[0]
del actors_dict[-1]
```

```python
actors_dict[0]
```

## 3. 배우 나이, 작품 활동 

```python
age_list = []
activation_list = []

for i in tqdm_notebook(actors_dict):
    driver2 = set_chrome_driver()
    driver2.get(i['Name_url'])
    html = driver2.page_source
    soup = BeautifulSoup(html)
    if soup.select('div.zvclcYLT.ocV8Dgty table.zQHAoFCY._ba832eed82a14e6b3bc9304cd9b2a6d0 div.HMXdAcwv a._0BBV4ix3') != []:
        age = soup.select('div.zvclcYLT.ocV8Dgty table.zQHAoFCY._ba832eed82a14e6b3bc9304cd9b2a6d0 div.HMXdAcwv a._0BBV4ix3')[0].text
    else:
        age = None
    age_list.append(age)
    
    if soup.select('div._6IV8B0dU') != []:
        activation = soup.select('div._6IV8B0dU')[1].text
    else:
        activation = None
    
    print(age)
    print(activation)
    activation_list.append(activation)
    #driver2.refresh()
    driver2.close()
```

```python
df = pd.DataFrame(actors_dict,
            columns=['Name', 'Name_url'])
```

```python
df['birth_year'] = age_list
df['activation_content'] = activation_list
```

```python
df.to_csv('Actor_namuwiki.csv', encoding='utf-8-sig', index=False)
```

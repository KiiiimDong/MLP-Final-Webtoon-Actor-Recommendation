## 데이터 로드

```python
actor_df = pd.read_csv('../version_1.0/text_data/Actor_keywords.csv', encoding='utf-8-sig')
actor_df.head()
```

```python
webtoon_df = pd.read_csv('../version_1.0/text_data/Webtoon_keywords.csv', encoding='utf-8-sig')
webtoon_df.head()
```

## 키워드 유사도 점수 Table 만들기

```python
# keyword 객체 생성
model_keyword = keyword.Keyword('../version_1.0/model_fasttext/fasttext.model')
```

```python
keyword_score_list = []

for i in tqdm(range(len(webtoon_df))):
    keyword_score_dict = dict()
    
    webtoon_title = webtoon_df['Title'][i]
    webtoon_character_name = webtoon_df['Name'][i]
    webtoon_character_5_keywords = webtoon_df['5_keywords'][i]
    
    keyword_score_dict['Title'] = webtoon_title
    keyword_score_dict['Name'] = webtoon_character_name
    
    for j in range(len(actor_df)):
        actor_name = actor_df['Name'][j]
        actor_age = actor_df['Age'][j]
        actor_sex = actor_df['Sex'][j]
        actor_5_keywords = actor_df['5_keywords'][j]
        
        keyword_score = model_keyword.keyword_similarity(webtoon_character_5_keywords, actor_5_keywords)
        
        keyword_score_dict[f'{actor_name}_{actor_age}_{actor_sex}'] = keyword_score
        
    keyword_score_list.append(keyword_score_dict)
```

```python
keyword_score_df = pd.DataFrame(keyword_score_list)
keyword_score_df
```

```python
# 저장
keyword_score_df.to_excel('../version_1.0/text_data/Keyword_score_table_3.xlsx', encoding='utf-8-sig', index=False)
```
